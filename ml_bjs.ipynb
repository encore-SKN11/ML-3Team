```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib
import warnings
warnings.filterwarnings('ignore')

font_path = 'C:\\Windows\\Fonts\\gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
matplotlib.rc('font',family=font)
```

```python
# 데이터 로드
df = pd.read_csv('./data/cancer_patient_data_sets.csv')
df
```

```python
# 상대적으로 연관이 없는 컬럼 삭제하기

df = df.drop(['index',
            'Patient Id',
            'Gender',
              'Fatigue', 
              'Weight Loss', 
              'Shortness of Breath', 
              'Wheezing', 
              'Swallowing Difficulty', 
              'Clubbing of Finger Nails', 
              'Frequent Cold', 
              'Dry Cough',
              'Snoring'],
              axis=1)

df['Level'] = df['Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
df
```

```python
# 입력-라벨 데이터 분리
# X = related_df.drop('Level', axis=1)
# y = related_df['Level']


X = df.drop('Level', axis=1)
y = df['Level']
y
```

```python
# 훈련-테스트 데이터 분리 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# # 표준화 
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)
```

```python
# 표준화 
from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_y.fit_transform(X_test)
```

### 랜덤 포레스트 분류 모델

```python
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트 분류 
rf_clf = RandomForestClassifier(n_estimators=50, max_depth=7, random_state=42)
```

```python
from sklearn.metrics import accuracy_score

# 학습
rf_clf.fit(X_train_scaled, y_train)

# 예측
y_pred_train = rf_clf.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train, y_pred_train)
print("학습 점수:", acc_score_train)

y_pred_test = rf_clf.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test, y_pred_test)
print("테스트 평가 점수:", acc_score_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, classification_report
report_rf = classification_report(y_test, y_pred_test)
print(report_rf)
```

### Logistic Regression 모델

```python
from sklearn.linear_model import LogisticRegression

# LogisticRegression 다중분류 모델 객체 생성
lr_clf = LogisticRegression(multi_class='multinomial', solver='lbfgs')

# 학습
lr_clf.fit(X_train_scaled, y_train)

# 예측 및 평가

y_pred_train = lr_clf.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train, y_pred_train)
print("학습 점수:", acc_score_train)

print(lr_clf.score(X_train_scaled, y_train), lr_clf.score(X_test_scaled, y_test))

y_pred_test = lr_clf.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test, y_pred_test)
print("테스트 평가 점수:", acc_score_test)

report_lr = classification_report(y_test, y_pred_test)
print(report_lr)
```

### KNN 모델

```python
from sklearn.neighbors import KNeighborsClassifier

# KneighborsClassifier 모델 객체 생성
knn = KNeighborsClassifier(n_neighbors=5)

# 학습
knn.fit(X_train, y_train)

# 예측 및 평가
y_pred_train = knn.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train, y_pred_train)
print("학습 점수:", acc_score_train)

y_pred_test = knn.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test, y_pred_test)
print("테스트 평가 점수:", acc_score_test)

report_knn = classification_report(y_test, y_pred_test)
print(report_knn)
```

### KNN 모델 GridSearchCV로 최적의 하이퍼파라미터 찾아서 진행

```python
from sklearn.model_selection import GridSearchCV 
# 하이퍼파라미터 후보 설정
param_grid_knn = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

# GridSearchCV 적용
knn = KNeighborsClassifier()
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring="accuracy", n_jobs=-1)
grid_search_knn.fit(X_train, y_train)

# 최적 모델 선택
best_knn = grid_search_knn.best_estimator_
print("KNN 최적 하이퍼파라미터:", grid_search_knn.best_params_)
```

```python
# 최적 모델로 예측
y_pred_knn = best_knn.predict(X_test)

# 평가
print("\n[KNN 평가 결과]")
print("정확도:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

### decision tree 모델 GridSearchCV로 최적의 하이퍼파라미터 찾아서 진행

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# 클래스 라벨 변환 (0부터 시작하도록) why? XGBoost는 다중분류에서 클래스를 0부터 시작해야하기 때문에 
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)  # 0부터 시작하는 정수로 변환
y_test_encoded = encoder.transform(y_test)

# XGBClassifier 모델 객체 생성
xgb_clf = XGBClassifier(
    objective="multi:softmax",  # 다중분류 (softmax 사용)
    eval_metric="logloss",
    learning_rate=0.1,
    n_estimators=100,
    max_depth=3,
    random_state=42
)

# 모델 학습
xgb_clf.fit(X_train_scaled, y_train_encoded)

# 예측 및 평가
y_pred_train = xgb_clf.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train_encoded, y_pred_train)
print("학습 점수:", acc_score_train)

y_pred_test = xgb_clf.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test_encoded, y_pred_test)
print("테스트 평가 점수:", acc_score_test)

report_xgb = classification_report(y_test_encoded, y_pred_test)
print(report_xgb)
```

### Soft Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# 클래스 라벨 변환 (0부터 시작하도록)
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# 개별 모델 정의
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
lr_clf = LogisticRegression(max_iter=1000, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
svc_clf = SVC(probability=True, random_state=42)  # 해결: probability=True 설정
dt_clf = DecisionTreeClassifier(random_state=42)
xgb_clf = XGBClassifier(
    objective="multi:softmax",
    eval_metric="logloss",
    learning_rate=0.1,
    n_estimators=100,
    max_depth=3,
    random_state=42
)

# 투표 기반 분류기 (VotingClassifier) 생성
voting_clf = VotingClassifier(
    estimators=[
        ('rf', rf_clf),
        ('lr', lr_clf),
        ('knn', knn),
        ('svc', svc_clf),  # 수정된 SVC
        ('dt', dt_clf),
        ('xgb', xgb_clf)
    ],
    voting='soft'  # 확률 기반 투표
)

# 모델 학습
voting_clf.fit(X_train_scaled, y_train_encoded)

# 예측 및 평가
y_pred_train = voting_clf.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train_encoded, y_pred_train)
print("학습 점수:", acc_score_train)

y_pred_test = voting_clf.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test_encoded, y_pred_test)
print("테스트 평가 점수:", acc_score_test)

report_voting = classification_report(y_test_encoded, y_pred_test)
print(report_voting)
```

### Hard Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# 클래스 라벨 변환 (0부터 시작하도록)
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# 개별 모델 정의
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
lr_clf = LogisticRegression(max_iter=1000, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
svc_clf = SVC(probability=True, random_state=42)  # 해결: probability=True 설정
dt_clf = DecisionTreeClassifier(random_state=42)
xgb_clf = XGBClassifier(
    objective="multi:softmax",
    eval_metric="logloss",
    learning_rate=0.1,
    n_estimators=100,
    max_depth=3,
    random_state=42
)

# 투표 기반 분류기 (VotingClassifier) 생성
voting_clf = VotingClassifier(
    estimators=[
        ('rf', rf_clf),
        ('lr', lr_clf),
        ('knn', knn),
        ('svc', svc_clf),  # 수정된 SVC
        ('dt', dt_clf),
        ('xgb', xgb_clf)
    ],
    voting='hard'  # 다수결 투표
)

# 모델 학습
voting_clf.fit(X_train_scaled, y_train_encoded)

# 예측 및 평가
y_pred_train = voting_clf.predict(X_train_scaled)
acc_score_train = accuracy_score(y_train_encoded, y_pred_train)
print("학습 점수:", acc_score_train)

y_pred_test = voting_clf.predict(X_test_scaled)
acc_score_test = accuracy_score(y_test_encoded, y_pred_test)
print("테스트 평가 점수:", acc_score_test)

report_voting = classification_report(y_test_encoded, y_pred_test)
print(report_voting)
```
