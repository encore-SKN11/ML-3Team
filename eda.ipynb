```python
# 상대적으로 연관이 없는 컬럼 삭제하기

related_df = df.drop(['index',
            'Patient Id',
            'Gender',
              'Fatigue', 
              'Weight Loss', 
              'Shortness of Breath', 
              'Wheezing', 
              'Swallowing Difficulty', 
              'Clubbing of Finger Nails', 
              'Frequent Cold', 
              'Dry Cough',
              'Snoring'],
              axis=1)

related_df['Level'] = related_df['Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
related_df
```

# 성별 폐암 환자 분포 시각화

```python
# 성별 폐암 환자 분포 시각화

# Gender 숫자 → 텍스트 변환
df['Gender'] = df['Gender'].replace({1: 'Male', 2: 'Famale'})

# 그래프 그리기
plt.figure(figsize=(6, 5))
ax = sns.countplot(x=df['Gender'], palette='pastel')

# 각 bar 위에 개수 추가
for bar in ax.patches:
    x = bar.get_x() + bar.get_width() / 2  # 막대의 중심
    y = bar.get_height()  # 막대 높이
    label = bar.get_x()  # Male, Female 라벨
    plt.text(x, y + 3, f"{int(y)}명", ha='center', fontsize=11, fontweight='bold', color='black')

plt.title("성별 폐암 환자 분포", fontsize=14, fontweight='bold')
plt.xlabel("성별", fontsize=12)
plt.ylabel("명 수", fontsize=12)
plt.show()
```

# 연령별 폐암 환자 분포 시각화

```python
# 연령별 폐암 환자 분포 시각화 
# 폐암 환자의 연령은 대부분 28~45세
plt.figure(figsize=(4, 6))
sns.boxplot(y=related_df['Age'], color='lightblue')

# Q1 (Box의 아래 경계) & Q3 (Box의 위 경계) 구하기
stat = related_df['Age'].describe()
box_min = stat['25%']  # Q1 (Box의 아래쪽 값)
box_max = stat['75%']  # Q3 (Box의 위쪽 값)
age_min = related_df['Age'].min()
age_max = related_df['Age'].max()

# 값 표시
plt.text(0, box_min - 2, f"Box Min: {box_min:.1f}세", ha='center', color='blue', fontsize=12)
plt.text(0, box_max + 2, f"Box Max: {box_max:.1f}세", ha='center', color='blue', fontsize=12)
plt.text(0, age_min - 2, f"min: {age_min}세", ha='center', color='red', fontsize=12)
plt.text(0, age_max + 1, f"max: {age_max}세", ha='center', color='red', fontsize=12)


plt.title("연령별 폐암 환자 분포", fontsize=14, fontweight='bold')
plt.ylabel("나이", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.text
plt.show()
```

# 각 컬럼별 시각화 (Boxplot)

```python
# 각 컬럼별 boxplot (age, level 제외)
fig, axes = plt.subplots(3, 4, figsize=(12, 10))  # 4x2 그리드로 배치
fig.suptitle("각 컬럼별 Boxplot", fontsize=16, fontweight='bold')

# 컬럼별로 boxplot 생성
for ax, col in zip(axes.flatten(), related_df.columns[1:13]):
    sns.boxplot(y=df[col], ax=ax, color='lightblue')

    # 평균값 계산 및 표시
    mean_value = related_df[col].mean()
    ax.axhline(mean_value, color='red', linestyle='--', linewidth=1)  # 평균선 추가
    ax.text(0, mean_value, f"{mean_value:.2f}", ha='center', va='bottom', 
            color='red', fontsize=10, fontweight='bold')  # 평균값 텍스트 추가

    ax.set_title(col, fontsize=12)  # 각 subplot 제목
    ax.set_xlabel("")  # x축 라벨 제거
    ax.set_ylabel("")  # y축 라벨 제거

# 레이아웃 조정
plt.tight_layout(rect=[0, 0, 1, 0.96])  # 제목 공간 확보
plt.show()
```

# 각 특성에 따른 폐암 Level 분포

```python
# 각 특성의 수준별 폐암 level 시각화
# 비만, 음주, 균형잡힌 식단, 먼지 알레르기, 유전적 요인, 간접흡연
# x축은 특성의 수준(1-8), y축은 특성의 수준별 level
```

```python
# 특성 리스트
features = ['Obesity', 'Alcohol use', 'Balanced Diet', 'Dust Allergy', 'Genetic Risk', 'Passive Smoker']

# 그래프 설정
fig, axes = plt.subplots(2, 3, figsize=(20, 14))  # 3x2 그리드로 배치
fig.suptitle("특성에 따른 Level 분포 (Count)", fontsize=16, fontweight='bold')

# 각 특성에 대해 반복적으로 그래프 그리기
for i, feature in enumerate(features):
    ax = axes.flatten()[i]  # 각 subplot에 해당하는 axes를 가져옴
    level_counts = related_df.groupby([feature, 'Level']).size().unstack(fill_value=0)
    
    # 꺾은선 그래프 그리기
    sns.lineplot(data=level_counts, x=level_counts.index, y=1, label='Level 1', marker='o', 
                 markerfacecolor='red', markeredgewidth=2, linewidth=2, color='red', ax=ax)
    sns.lineplot(data=level_counts, x=level_counts.index, y=2, label='Level 2', marker='s', 
                 markerfacecolor='blue', markeredgewidth=2, linewidth=2, color='blue', ax=ax)
    sns.lineplot(data=level_counts, x=level_counts.index, y=3, label='Level 3', marker='^', 
                 markerfacecolor='green', markeredgewidth=2, linewidth=2, color='green', ax=ax)
    
    # 제목과 레이블 추가
    ax.set_title(f'{feature}에 따른 Level 분포', fontsize=14, fontweight='bold')
    ax.set_xlabel(feature, fontsize=12)
    ax.set_ylabel('Level Count', fontsize=12)
    
    # 범례 추가
    ax.legend(title='Level', loc='upper left', fontsize=10, frameon=True, framealpha=0.7)

    # 그리드 선 추가
    ax.grid(True, linestyle='--', alpha=0.7)

    # x축 레이블 회전
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=10)

# 레이아웃 조정
plt.tight_layout(rect=[0, 0, 1, 0.96])  # 제목 공간 확보
plt.show()
```

```python
level_counts = related_df.groupby(['Air Pollution', 'Level']).size().unstack(fill_value=0)

# 그래프 그리기
plt.figure(figsize=(8, 6))
sns.lineplot(data=level_counts, x=level_counts.index, y=1, label='Level 1', marker='o', color='red', linewidth=2)
sns.lineplot(data=level_counts, x=level_counts.index, y=2, label='Level 2', marker='s', color='blue', linewidth=2)
sns.lineplot(data=level_counts, x=level_counts.index, y=3, label='Level 3', marker='^', color='green', linewidth=2)

# 그래프 꾸미기
plt.title("Air Pollution에 따른 Level 분포", fontsize=14, fontweight='bold')
plt.xlabel("Air Pollution", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Level")
plt.grid(axis='y', linestyle="--", alpha=0.7)

plt.show()
```
